```python
#reading the csv file and handling the null values
import pandas as pd
df = pd.read_csv('orders.csv',na_values=['Not Available','unknown'])
df['Ship Mode'].unique()
df.head()
```

```python
#renaming the column names
#df.rename(columns={'Order Id':'order_id','Order Date':'order_date','Ship Mode':'ship_mode','Segment':'segment','Country':'country','City':'city','State':'state','Postal Code':'postal_code','Region':'region','Category':'category','Sub Category':'sub_category','Product Id':'product_id','cost price':'cost_price','List Price':'list_price','Quantity':'quantity','Discount Percent':'discount_percent'})
#another way
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ","_")
df.columns
df.head(10)
```

```python
#creating the required new columns like discount, selling price, profit
df['discount'] = (df['list_price'] * df['discount_percent']) / 100
df['selling_price'] = df['list_price'] - df['discount']
df['profit'] = df['selling_price'] - df['cost_price']
df.head(10)
```

```python
#convert the data type of order_date of object datetime to normal datetime
df['order_date'] = pd.to_datetime(df['order_date'],format="%Y-%m-%d")
df.dtypes
```

```python
#removing the columns cost price, list price and discount percent
df.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
#after running the above command once it will show error because deleted columns are removed from original data, so after its operation commenting it. 
df.head(10)
```

```python
#creating table with actual column names and types from your CSV
# ensure to truncate the table df_orders_table in mysql when execute again or it will append the repeated values again and again so that result will not be coorect. 
create_table_query = """
CREATE TABLE IF NOT EXISTS df_orders_table (
    order_id INT,
    order_date DATE,
    ship_mode VARCHAR(20),
    segment VARCHAR(20),
    country VARCHAR(20),
    city VARCHAR(20),
    state VARCHAR(20),
    postal_code VARCHAR(20),
    region VARCHAR(20),
    category VARCHAR(20),
    sub_category VARCHAR(20),
    product_id VARCHAR(50),
    quantity INT,
    discount DECIMAL(7,2),
    selling_price DECIMAL(7,2),
    profit DECIMAL(7,2)
)
"""
Connection.execute(create_table_query)
print("Table Created")
```

```python
df['order_date'] = df['order_date'].dt.strftime('%Y-%m-%d')  # avoids timestamp errors
```

```python
#inserting into db
data = df[[
    'order_id',
    'order_date',
    'ship_mode',
    'segment',
    'country',
    'city',
    'state',
    'postal_code',
    'region',
    'category',
    'sub_category',
    'product_id',  
    'quantity',
    'discount',
    'selling_price',
    'profit'
]].values.tolist()

insert_query = """
INSERT INTO df_orders_table (
    order_id,
    order_date,
    ship_mode,
    segment,
    country,
    city,
    state,
    postal_code,
    region,
    category,
    sub_category,
    product_id,
    quantity,
    discount,
    selling_price,
    profit
) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""

Connection.executemany(insert_query, data)
db.commit()
Connection.close()
db.close()
print("CSV data inserted into MySQL.")
```
